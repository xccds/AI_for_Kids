### 什么是人工智能
人工智能（Artificial Intelligence,AI）是人类最伟大的梦想之一。而人们对人工智能的了解恐怕还是主要来自于电影中的角色，例如在“2001太空漫游”电影中的AI电脑管家HAL，“终结者”系列电影中的杀手机器人，以及“钢铁侠”电影中的战衣AI。电影中的这些人工智能都具备一些共同特征：能力异常强大，和人一样会思考，甚至会有感情。但是现实世界中的AI却相差甚远，不过它们已经在我们身边发挥着作用。例如可以自动帮我们打扫房间的扫地机器人，可以和人愉快聊天的网络机器人等等。

到底什么是人工智能，这个概念也是不断发展的。最早科学家们认为人工智能需要像人一样思考和行动，完全的模仿人类。但是，人类建造的飞机最终能飞上天，依赖是并不是对鸟的模仿，而是对空气动力学的研究。现代的观点则认为，人工智能需要能合理的行动，它能够自主的操作、感知环境、长期持续、适应变化并实
现最佳期望结果，也就是最优的完成一个即定的目标。所以人工智能需要通过圆满完成任务来检验其能力。



### 为什么写这本书
近年以来，随着数据的丰富，算法和算力的加强，以深度学习、强化学习为代表的AI技术迅速崛起，《麻省理工学院技术评论》杂志将深度学习列为2013年十大突破性技术之首，将强化学习列为2017年十大突破性技术之首。2017年，谷歌公司开发了会下围棋的人工智能Alpha Go和Alpha Zero，战胜了李世石、柯洁在内的多名人类围棋大师。同年，中国国务院发布了《新一代人工智能发展规划》，强调“实验全民智能教育项目，在中小学阶段设置人工智能相关课程”。由此我们感受到，人工智能的时代已经到来。中国的青少年，需要学会用AI的语言来理解世界，用AI的思想来创造世界。

人工智能领域，博大精深，理论繁复，工程框架众多。从现有的出版情况来看，不论是图书还是论文，已经非常丰富了。但是大部分是针对的业界或学界的专业人士，而针对青少年的、入门级的读物并不算多。因此笔者希望能写一本门槛较低，学习曲线平缓，生动形象不枯燥的人工智能普及读物。

### 本书的特点
和武侠世界中类似，学习人工智能也需要同时修炼内功和外功，内功是对于各种理论知识点的理解，外功是对于编程工具的运用。
本书的特点是三个关键词来概括：
- 表达生动：对于理论知识点的介绍，本书主要通过通俗易懂的语言和图表来讲解，回避过多的公式。
- 代码为主：python代码在表达能力上比公式更为具体，而比语言更为精确，所以本书有大量代码，不仅帮助理解知识点，也加强读者的编程能力。
- 知识完备：人工智能领域需要的知识点相当多，本书选择了其中最为重要最为基础的知识点，而且均从基础讲起，包括了重要的机器学习入门知识，编程算法，不需要其它的资料作为预备读物。


### 本书的内容安排
本书分为三大部分，共24节课程内容：
- 第一部分包括了第2课到第6课，主要讲解基础编程知识，学习Python语言特性，同时也学习计算机经典算法，例如快速排序算法、广度优先算法、贪心算法等。
- 第二部分包括了第7课到第11课，主要讲解如何来处理数据，如何分析和观察数据，同时介绍了重要的随机模拟和优化算法，这是机器学习的先修内容。
- 第三部分包括了第12课到第23课，主要讲解机器学习的整体框架，重点介绍了以神经网络为主的若干机器学习算法，并进一步拓展到了深度学习和强化学习的入门知识。
- 此外，第1课介绍了运行本书代码所需要的环境安装，第24课介绍了人工智能的发展领域，以及继续学习研究的参考资源。

### 适合本书的读者
- 开设人工智能课程的中小学老师同学
- 有一定的编程热情和实践环境青少年朋友
- 希望了解人工智能领域的成人家长朋友

### 本书的学习方法
- 按章节顺序学习，因为后面的知识点需要用到前面内容打基础。
- 一定要在计算机上写代码操作，因为人工智能领域的技术含量很高，特别是用到导数、矩阵等高等数学的地方，远远超出了中小学的学习范畴，这些难点本书是通过代码的方式作讲解，而回避了理论公式。所以需要通过编写代码，运行代码，来理解代码背后的知识，由外功修内功。
- 反复练习，学习就是模仿和重复，咬定青山不放松。
- 有条件的情况下，可以成人家长和小孩一起学习，一起成长。
