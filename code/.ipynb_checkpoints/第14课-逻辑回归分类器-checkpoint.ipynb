### 1. 什么是逻辑回归分类器

在上一课中，我们学习了感知机分类器，它的核心思想有两部分，一部分是如何处理输入数据，感知机是以加权求和的方式进行汇总，然后用一个阶梯函数进行最终输出。另一部分是如何调整得到权重，感知机是根据真实结果和预测结果的距离来进行修改。逻辑回归分类器和感知机是类似的，它们的相同点在于，都是用加权求和的方式来汇总输入数据。区别有三个地方：
- 如果是二分类问题，逻辑回归的真实标签一般编码为0，1方式，而不是感知机的-1，+1的方式。
- 感知机使用阶梯函数进行最终的输出转换，这个转换功能又称之为激活，逻辑回归使用的是一个更为平滑的激活函数，叫做sigmoid函数。
- 逻辑回归的权重修正公式有不同的推导过程，它是从一个交叉熵损失函数推导出来的。

逻辑回归会对数据数据做一个汇总处理，这种处理和感知机是一样的。
$$z = w\cdot x+b$$
也可以写成下面的形式
$$z = w_0x_0 + w_1x_1 + ... + w_mx_m = \sum_{i=0}^nw_ix_i$$

然后对于这个汇总的量z，我们用一个不同的激活函数进行转换，转换的目的是将输出转成0到1之间的值，激活函数就是sigmoid函数，它的具体形式是如下：
$$\displaystyle  f(z) = \frac{1}{1+e^{-z}}$$

因此，整体的公式可以写成如下形式，它也就是逻辑回归的预测值。 $$\displaystyle f(x) = \frac{1}{1+e^{-(w\cdot x+b)}}$$

我们可以将sigmoid的函数形式画出来，直观的感觉一下。

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)

plt.plot(z, phi_z)
plt.axvline(0.0, color='k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')

# y axis ticks and gridline
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)

plt.tight_layout()
```

可以看到，z值越大，输出结果越接近于1，z值越小，输出结果越接近于0，所以可以认为f(x)是输出的预测结果为1的一个概率值。

### 2. 什么是损失函数
上面我们定义了如何根据输入数据来得到预测值，那么接下来我们需要定义预测值和真实标签之间的距离公式。

$$ LOSS={\begin{cases}-log(f(x))&{\text{if }}y=1\\-log(1-f(x))&{\text{if }}y=0\end{cases}}$$

为了理解这个距离公式，我们将它画在图上。

```python
def cost_1(z):
    return - np.log(sigmoid(z))
                   
def cost_0(z):
    return - np.log(1 - sigmoid(z))

z = np.arange(-10, 10, 0.1)
phi_z = sigmoid(z)

c1 = [cost_1(x) for x in z]
plt.plot(phi_z, c1, label='LOSS if y=1')

c0 = [cost_0(x) for x in z]
plt.plot(phi_z, c0, linestyle='--', label='LOSS if y=0')

plt.ylim(0.0, 5.1)
plt.xlim([0, 1])
plt.xlabel('f(x)')
plt.ylabel('LOSS')
plt.legend(loc='best')
plt.tight_layout()
```

从图上可以看到，当真实标签取1的时候，如果预测值越接近1，这个距离就越小，反之越大。当真实标签取0的时候，如果预测值越接近于0，距离越小，反之越大。所以这个公式反映了y和f(x)之间的一种距离度量，就类似于它们之间做了一个减法。一个样本的损失函数也可以写成一个公式。

$$ LOSS = -(y log(f(x)) + (1-y) log(1-f(x))) $$

这个距离公式在机器学习中一般称为损失函数，不同的算法会有不同定义的损失函数，这种损失函数称之为交叉熵损失函数。损失函数定义好之后，我们就可以基于最优化的思路，找到一个最优的权重，使之对应的损失函数最小。这就是定义损失函数的目的。

下面我们还是用著名的鸢尾花数据集来实验一下。

```python
import pandas as pd

df = pd.read_csv('../data/iris.csv')
df.head()
```

```python
y = df.iloc[0:100, 4].values
y = np.where(y == 'Setosa', 0, 1)

# extract sepal length and petal length
X = df.iloc[0:100, [0, 2]].values
```

注意使用逻辑回归的时候最好是有一个数据预处理，就是标准化。标准化就是先计算出变量的均值和标准差，然后(每个样本的观测值-均值)/标准差

```python
X_std = np.copy(X)
X_std[:,0] = (X[:,0] - X[:,0].mean()) / X[:,0].std()
X_std[:,1] = (X[:,1] - X[:,1].mean()) / X[:,1].std()
```

```python
# use Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_std, y)
```

逻辑回归分类器训练好之后，可以输入一个新的样本，来看一下预测如何

```python
new_x = np.array([[5.3, 1.2]])
lr.predict_proba(new_x)
```

和感知机一样，我们也能看到计算的对应权重，以决策边界。

```python
w1, w2 = lr.coef_[0]
print("w1:",w1,"\n""w2:",w2)
```

```python
b = lr.intercept_[0]
print("b:",b)
```

```python
a = -w1/w2
line_x = np.linspace(-2,2,100)
line_y = line_x*a - b
```

```python
plt.scatter(X_std[:50, 0], X_std[:50, 1],
            color='red', marker='o', label='setosa')
plt.scatter(X_std[50:100, 0], X_std[50:100, 1],
            color='blue', marker='x', label='versicolor')
plt.plot(line_x,line_y)
plt.xlabel('petal length [cm]')
plt.ylabel('sepal length [cm]')
plt.legend(loc='upper left')

plt.tight_layout()
```

### 3. 逻辑回归的权重如何训练的

```python
from IPython.display import Image
Image(filename='../figures/03_03.png', width=600)
```

和感知机类似，那些权重W也是通过反复试错迭代的思路学习出来的。具体的步骤可以是如下的几个步骤：

1.先给W一个随机的赋值，

2.计算给定W下的模型预测值

3.计算模型预测值和真实目标值之间的差距，我们叫做损失函数

4.根据损失函数来对W求导，根据梯度下降算法来修正W

每次迭代权重向量以如下方式修正更新：
$$ w:= w + eta * (y-f(x))*x$$


```python
class LogisticRegression(object):
    def __init__(self, eta=0.01, n_iter=50):
        self.eta = eta
        self.n_iter = n_iter

    def fit(self, X, y):
        self.w_ = np.zeros(1 + X.shape[1])
        self.cost_ = []       
        for i in range(self.n_iter):
            y_val = self.activation(X)
            errors = (y - y_val)
            neg_grad = X.T.dot(errors)
            self.w_[1:] += self.eta * neg_grad
            self.w_[0] += self.eta * errors.sum()
            self.cost_.append(self._logit_cost(y, self.activation(X)))
        return self

    def _logit_cost(self, y, y_val):
        logit = -y.dot(np.log(y_val)) - ((1 - y).dot(np.log(1 - y_val)))
        return logit
    
    def _sigmoid(self, z):
        return 1.0 / (1.0 + np.exp(-z))
    
    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        z = self.net_input(X)
        return self._sigmoid(z)
    
    def predict_proba(self, X):
        return activation(X)

    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, 0)
```

```python

lr = LogisticRegression(n_iter=500, eta=0.02).fit(X_std, y)
plt.plot(range(1, len(lr.cost_) + 1), np.log10(lr.cost_))
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.title('Logistic Regression - Learning rate 0.02')

plt.tight_layout()
```

### 4. 分类器的评估方法

对于一个分类器的结果，或者说模型的结果，我们需要评价它的性能怎么样，是不是足够的好。下面我们来介绍几种常见的评估方法。但在介绍评估方法前先介绍两个重要的概念，就是训练集和检验集。
- 训练集：用来模型训练的数据
- 检验集：用来观察模型性能的数据

类似于我们在数学课上做练习和测验一样，老师会给同学们讲一些练习题，这些习题是用来帮助同学理解知识的。然后为了检查同学们有没有真正的学到知识，会用另外一些单独的习题来做测试。这种测试用的题目和讲解用的题目应该不会是完全一样的，否则大家只需要死记硬背就可以了。那么和这些习题类似的作法，我们在训练一个模型的时候使用一批数据，在检查它是否真正学到的时候，应该用另一批数据，所以这两批数据目的不同，这也就是说为什么要区分训练集和检验集的原因了。

在sklean的模块中有一个函数可以帮助我们来划分两个数据集，那就是train_test_split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
         X_std, y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_predict = lr.predict(X_test)
y_predict_prob = lr.predict_proba(X_test)
```

```python
sklearn.metrics.accuracy_score(y_test,y_predict)
```

### 练习
我们来练习一下，损失函数是如何来计算的，假如有四个样本，它们的真实标签是[0, 0, 1, 1]，而预测出来1的概率值分别是[0.1, 0.2, 0.7, 0.99]，那么这组数据的损失是多少呢？

根据损失函数的定义
$$ LOSS = -(y log(f(x)) + (1-y) log(1-f(x))) $$

```python
def logit_cost(y, p):
    logit = -y.dot(np.log(p)) - ((1 - y).dot(np.log(1 - p)))
    return logit/len(y)
```

```python
y_true = np.array([0, 0, 1, 1])
y_pred = np.array([0.1, 0.2, 0.7, 0.99])
```

```python
logit_cost(y_true,y_pred)
```

我们也可以直接从sklearn中的metrics模块中获取现成的损失函数

```python
from sklearn.metrics import log_loss
log_loss(y_true, y_pred)
```
