### 1. 什么是机器学习

再谈机器学习之前，我们先思考一下，什么是学习？人类是如何学习的？人类的学习是理解知识，解决问题，人类的学习过程是从过去的经验中归纳规律，从而推运用到类似的领域。类别到机器学习，就是要让计算机能独立的总结规律，完成特定的任务。

在之前的章节例子中，我们为了完成某个任务，通常会针对这种任务设计一些方法，以达到目标，例如对一组数据做排序。但是对于一些较为复杂的任务，人工的设计会非常麻烦，而且无从下手。例如我们想设计一套算法来识别不同图片是否属于猫或者狗，一种可能的方法是通过人工，搜集尽可能多的猫狗照片，然后观察它们的区别，提取一些人工的经验规则，例如有胡子的动物就判断是一只猫，当然这类规则不一定完全准确，而且如何让计算机在图片中找到胡子也是很困难的。所以基于人工总结规则（专家系统）的方法难以保证完备和准确性，成本高而且耗时高。

那么可以考虑另一种思路，如果手头上有大量和猫狗有关的照片数据，而且每张照片都有人看过，而且标注过是猫还是狗，那么可以直接就图片和标注一起喂给计算机，让计算机自己去找到判断的规则，这就是典型的从数据中去学习规律的思路，也是机器学习重要思想。

所以机器学习应用的场景是有两个条件，一个是人工直接设计规则非常复杂，另一个是有大量数据可以直接应用。

机器学习可以大致分为三种：
- 有监督学习
- 无监督学习
- 强化学习

### 2. 有监督学习

有监督学习是指像前述的猫狗分类的问题，有人类专家标注的信息作为老师，来监督计算机学习的过程，对于每张照片，计算机会有一个初始的判断，如果计算机的判断和标注的一致，可以认为是判断正确的，如果不一致，则认为判断是错误的，就需要对这个判断进行修正优化。从过程上看起来并不复杂，但涉及到几个概念需要在下面澄清。

- 输入(input)：输入给计算机的信息，通常我们用X表示，上例的输入就是一张图片
- 目标(target)：人工专家定义的标注，通常用Y表示，上例中每张图片都会有一个标注，注明是猫或是狗，这种标注通常会转换成0-1表示。
- 预测(predict)：计算机需要判断预测的结果，上例的预测输出就是预测图片的类别或是预测的概率值。
- 模型(model)：计算机从输入到预测的判断逻辑，模型也可以认为是一种函数，函数输入，输出预测。最简单的函数就是一条直线，函数是有参数的，例如直线的斜率。
- 损失(loss)：如果人类专家的标注和模型的预测不一致，就认为模型还不够好，需要一种量化方法来计算这种不够好的程度，这种量化计算的结果称为损失，例如直接将标注值减去预测值。
- 优化(optimize)：一个良好的模型需要有较小的损失，所以机器学习问题也可以看成是一个最优化的问题，优化的目标是损失，可以调整的是模型的参数。

下面是一个具体的示例，来说明如何使python现在的模块来进行一个有监督学习的机器学习训练。

我们现在手头的问题是，给你一些关于花卉的数据，如何来判断这种花的种类。更具体点，是鸢尾花的种类。鸢【音：yuān】尾花（Iris）是单子叶百合目花卉，是一种比较常见的花，可能不经意间你就能在某个公园里碰见它。鸢尾花数据集可能是模式识别、机器学习等领域里被使用最多的一个数据集了，很多教材用这份数据来做案例，很多工具都会自带这些数据集，而且说学术界很多论文也应用这份数据做实验，可见这份数据的重要意义。

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from IPython.display import Image
Image(filename='iris.jpg', width=600)
```

鸢尾花数据集最初由Edgar Anderson 测量得到，而后在著名的统计学家和生物学家R.A Fisher于1936年发表的文章「The use of multiple measurements in taxonomic problems」中被使用，从此而被众人所知，尤其是在机器学习这个领域。数据中的鸢尾花记录结果是在加拿大加斯帕半岛上，于同一天的同一个时间段，使用相同的测量仪器，在相同的牧场上由同一个人测量出来的。这是一份有着70年历史的数据，虽然老，但是却很经典，详细数据集可以在UCI数据库中找到。

鸢尾花数据集共收集了三类鸢尾花，即Setosa鸢尾花、Versicolour鸢尾花和Virginica鸢尾花，每一类鸢尾花收集了50条样本记录，共计150条。
数据集包括4个属性，分别为花萼的长、花萼的宽、花瓣的长和花瓣的宽。对花瓣我们可能比较熟悉，花萼是什么呢？花萼是花冠外面的绿色被叶，在花尚未开放时，保护着花蕾。四个属性的单位都是cm，属于数值变量，四个属性均不存在缺失值的情况。

python中最常用的机器学习模块就是sklearn，它就包括了iris这个数据集，让我们先把这个数据读入进来，看一看。

```python
from sklearn.datasets import load_iris
from sklearn import linear_model
import numpy as np
```

```python
iris = load_iris()
```

```python
iris.data.shape
```

读入的数据会保存在iris中，它的四个属性数据或者说特征输入存在data中，它其实就是Numpy的数据格式，其行列数分别为150行，4列。下面我们可以将前五行观察样本打印出来看看。

```python
iris.data[:5]
```

iris中的目标变量存放在target中，我们也可以打印前五个样本看一下，同时也可以使用unique函数，将所有target可能的取值打印出来，可以看到三种不同的花对应了0，1，2三种数字。

```python
iris.target[:5]
```

```python
np.unique(iris.target)
```

```python
X = iris.data
Y = iris.target
```

在这个例子中，iris.data的四列数值就是机器学习中的输入，我们用X来表示，iris.target的那一列数值就是目标，我们用Y来表示。一个有监督学习的任务，就是给出了输入和目标，需要建立模型。下面我们用一个简单的线性模型来看一下怎么做。

```python
clf = linear_model.SGDClassifier(max_iter=10)
```

使用之前加载的linear_model模块中的SGDClassifier算法，具体原理等到下个章节再理解，这里只需要理解使用上的方式，就是调用这个函数，并在函数中定义一些训练的参数。这里定义的参数max_iter表示训练多少个迭代。定义好我们使用的建模算法之后，这个模型保存在clf中，相当于一个备用的状态了。然后使用fit方法，来将数据给到算法进行训练建模。

```python
clf.fit(X,Y)
```

最后可以使用模型的predict方法来对一些输入做出相应的预测

```python
clf.predict(X)[:5]
```

小结：有监督学习需要有标签来让算法去训练，标签就像老师一样起到了监督的作用，所以称之为有监督学习，学习到的结果是模型参数。

### 3. 无监督学习

无监督学习是指人类专家的标注并不存在，计算机自己从输入X中找到某些规律，典型的无监督学习方法就是聚类。聚类是没有标注的，它的目标是将若干距离非常接近的数据聚集在一起，形成群组。如何定义数据间的距离，如何构成群组，这是聚类算法需要解决的问题。

我们还是利用前面熟悉的iris数据集，假设这个数据集只知道输入X，并不知道目标Y，我们来对X做一些聚类的计算。

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)

y_km = km.fit_predict(X)
```

和前面做的分类模型是类似的，我们从cluster模块中读入kmeans算法，然后定义算法的参数，再将模型使用fit_predict方法结合X数据进行训练。注意这里训练的时候我们只使用了X，没有使用Y的信息。训练的输出保存在y_km中。

```python
plt.scatter(X[y_km==0,0], 
            X[y_km==0,1], 
            s=50, 
            c='lightgreen', 
            marker='s', 
            label='cluster 1')
plt.scatter(X[y_km==1,0], 
            X[y_km==1,1], 
            s=50, 
            c='orange', 
            marker='o', 
            label='cluster 2')
plt.scatter(X[y_km==2,0], 
            X[y_km==2,1], 
            s=50, 
            c='lightblue', 
            marker='v', 
            label='cluster 3')
plt.scatter(km.cluster_centers_[:,0], 
            km.cluster_centers_[:,1], 
            s=250, 
            marker='*', 
            c='red', 
            label='centroids')
plt.legend()
plt.show();
```

小结：无监督学习的情况是没有标签，所以需要依赖算法自己学习到数据中的规律或模式。

### 12.4 强化学习

可以从一个生活中的例子来直觉上理解强化学习，如果我们刚喂养了一只小狗，希望能训练它，因为它本身不懂人的语言，所以没法上课教育它，但是可以对它的行动进行矫正。如果它做出了对的行为，我们就进行一次奖励，如果它做出了错的行为，我们就进行一次惩罚。很快它就能明白应该去做哪些行动。其实马戏团中的很多动物正是被强化学习的思路所训练的。

回到机器学习领域，从前面的内容我们了解到，如果每个样本都可以有一个目标值，或者标签来对应的话，我们可以通过有监督学习，来训练一个模型，这个模型学会的是如何从一个输入到一个标签的输出；这里标签的作用就好比一个标准答案，对模型的结果进行监督修正，如果每个样本都没有目标值，我们只能通过无监督学习，通过数据本身来学习内在的结构。在现实世界中，存在某些介于二者之间的情况，例如说将一只老鼠放入一个迷宫之中。假设迷宫很小，只有4*4的格式这么大，有一个起始点，和一个出口，

如果是用有监督方法来训练，那么需要在迷宫中的每个格子上，都有给一个标签，确定下一步行走的方向。有时候很难得到这种指导信息。在强化学习中，每个行为不会直接得到指导，但是一系列的行为之后，可以得到指导，例如当老鼠到达了出口时候给予一次奖励，通过奖励使它明白之前的一系列行动中，哪些行动可能是正确的，哪些可能是无用的。经过多次试错，让它强化了正确的行动。这种方法我们叫做强化学习。

所以强化学习就是让计算机实现从一开始什么都不懂, 通过不断地尝试, 从错误中学习, 最后找到规律, 学会了达到目的的方法. 这就是一个完整的强化学习过程. 实际中的强化学习例子有很多. 比如近年前最有名的 Alpha go, 机器头一次在围棋场上战胜人类高手, 让计算机自己学着玩经典游戏。

### 本课小结：


- 机器学习是一类从数据中学习的方法，让计算机模拟或实现人类的一些学习行为，以自行判断和行动。 
- 机器学习有几种不同风格的大类。 其中最流行的是“有监督学习”，有监督学习的训练需要提供一组输入对应的标签，也就是正确答案。你提供给监督学习的算法样本越多，其分析新数据的能力就越精确。 而这也是监督学习的主要挑战。 创建带有标记样本的大数据非常耗时，同时也需要大量的人力。 在“无监督学习”中，机器学习的另一个分支，则没有标签。 换句话说，你提供输入，但不提供输出。 该算法对未标记的数据进行整理并找出模式。
- 另一个热门的领域是“强化学习”。在强化学习中，没有直接提供给机器所有的行动对应的正确答案，但会提供在一串行为后的回报，机器学着如何在环境中拿到高分, 表现出优秀的成绩. 而这些成绩背后却是他所付出的辛苦劳动, 不断的试错, 不断地尝试, 累积经验, 学习经验。
